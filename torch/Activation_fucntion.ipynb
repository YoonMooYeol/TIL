활성화 함수(Activation Function)는 인공 신경망에서 뉴런의 출력을 결정하는 함수입니다. 
쉽게 말해, 뉴런이 입력 신호를 받아 어떤 값을 출력할지 정하는 역할을 합니다.

# 활성화 함수가 필요한 이유

비선형성 도입: 활성화 함수는 신경망에 비선형성을 도입하여 복잡한 패턴을 학습할 수 있도록 합니다. 만약 활성화 함수가 없다면, 신경망은 여러 개의 선형 레이어를 쌓은 것과 동일하게 되어 선형 모델의 한계를 벗어날 수 없습니다.
출력 값 조정: 활성화 함수는 뉴런의 출력 값을 특정 범위로 제한하거나 특정 조건에 따라 값을 변환하여 학습 과정을 안정화하고 효율성을 높입니다.
주요 활성화 함수 종류

Sigmoid: 출력 값을 0과 1 사이로 제한합니다. 주로 이진 분류 문제에 사용됩니다.
장점: 출력 값이 확률로 해석될 수 있습니다.
단점: 기울기 소실 문제가 발생할 수 있습니다.
ReLU (Rectified Linear Unit): 입력 값이 0보다 작으면 0, 0보다 크면 입력 값을 그대로 출력합니다.
장점: 계산이 간단하고 학습 속도가 빠릅니다.
단점: Dying ReLU 문제가 발생할 수 있습니다.
Tanh (Hyperbolic Tangent): 출력 값을 -1과 1 사이로 제한합니다.
장점: Sigmoid 함수보다 기울기 소실 문제가 덜 발생합니다.
단점: ReLU 함수보다 계산량이 많습니다.
Softmax: 여러 개의 출력 값을 확률 분포로 변환합니다. 주로 다중 클래스 분류 문제에 사용됩니다.
활성화 함수 선택 기준

활성화 함수는 문제의 유형, 데이터의 특징, 신경망의 구조 등을 고려하여 선택해야 합니다.

이진 분류: Sigmoid
다중 클래스 분류: Softmax
회귀: ReLU, Tanh 등
은닉 레이어: ReLU, Tanh 등


```python
# 파이토치에서 활성화 함수 사용 예시
import torch.nn as nn

# ReLU 활성화 함수
relu = nn.ReLU()
output = relu(input_data)

# Sigmoid 활성화 함수
sigmoid = nn.Sigmoid()
output = sigmoid(input_data)

# Softmax 활성화 함수
softmax = nn.Softmax(dim=1)  # dim=1은 일반적으로 사용되는 값입니다. 필요에 따라 변경 가능
output = softmax(input_data)

# Tanh 활성화 함수
tanh = nn.Tanh()
output = tanh(input_data)

# 이처럼 파이토치에서는 torch.nn 모듈을 통해 다양한 활성화 함수를 제공합니다.
```

# 활성화 함수 설명

이 코드는 파이토치에서 제공하는 다양한 활성화 함수를 사용하는 예시입니다.

1. **ReLU (Rectified Linear Unit) 활성화 함수**:
    - `nn.ReLU()`를 사용하여 ReLU 활성화 함수를 정의합니다.
    - ReLU는 입력이 0보다 크면 그대로 출력하고, 0보다 작으면 0을 출력합니다.
    - 주로 딥러닝 모델에서 비선형성을 추가하기 위해 사용됩니다.

2. **Sigmoid 활성화 함수**:
    - `nn.Sigmoid()`를 사용하여 Sigmoid 활성화 함수를 정의합니다.
    - Sigmoid 함수는 입력 값을 0과 1 사이의 값으로 변환합니다.
    - 주로 이진 분류 문제에서 출력층에 사용됩니다.

3. **Softmax 활성화 함수**:
    - `nn.Softmax(dim=1)`를 사용하여 Softmax 활성화 함수를 정의합니다.
    - Softmax 함수는 입력 벡터를 확률 분포로 변환합니다.
    - 주로 다중 클래스 분류 문제에서 출력층에 사용됩니다.
    - `dim=1`은 소프트맥스가 적용될 차원을 지정합니다.

4. **Tanh (Hyperbolic Tangent) 활성화 함수**:
    - `nn.Tanh()`를 사용하여 Tanh 활성화 함수를 정의합니다.
    - Tanh 함수는 입력 값을 -1과 1 사이의 값으로 변환합니다.
    - Sigmoid 함수와 유사하지만 출력 범위가 -1에서 1로 확장됩니다.

이처럼 파이토치에서는 `torch.nn` 모듈을 통해 다양한 활성화 함수를 제공합니다.
